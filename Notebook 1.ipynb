### Calculate a weighted rating (popularity-based)

WR = (v / (v+m))*R + (m / (v+m))*C     # the formula is from IMDB,

v - number of votes for a movie,

m - minimum number of votes required,

R - average rating of a movie, 

C - average rating across all movies

```
m = movies['vote_count'].quantile(0.90)
m
```

```
C = movies['vote_average'].mean()
C
```

```
movies['weighted_rating'] = (movies['vote_count']/(movies['vote_count']+m))*movies['vote_average'] + (m/(movies['vote_count']+m))*C

movies.head()
```

```
def weighted_rating(df, m=m, C=C):
    v = df['vote_count']
    R = df['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```
movies['weighted_rating'] = movies.apply(weighted_rating, axis=1)
movies
```

```
movies_filtered = movies.copy()
movies_filtered = movies_filtered.sort_values('weighted_rating', ascending=False)[['title', 'vote_count', 'vote_average', 'weighted_rating']].head(20)

dict___ = movies_filtered.to_dict('records')
dict___
```
